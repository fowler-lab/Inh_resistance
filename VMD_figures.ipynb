```python
import numpy as np
import MDAnalysis as mda
import MDAnalysis.analysis.align
import pandas as pd
pd.options.display.max_columns = 999
from MDAnalysis.tests.datafiles import PSF, DCD

KatG = mda.Universe("7ag8.pdb")
KatG_depth = '7ag8_depth.pdb'
KatG_hem_dist = '7ag8_hem_dist.pdb'
KatG_site1_dist = '7ag8_site1_dist.pdb'
KatG_snap2 = '7ag8_snap2.pdb'
KatG_site2_dist = '7ag8_site2_dist.pdb'
KatG_tempfactor = '7ag8_tempfactor.pdb'

titles = [KatG_depth,KatG_hem_dist,KatG_site1_dist,KatG_snap2,\
          KatG_site2_dist,KatG_tempfactor]
column = ['Depth','Hem_dist','Distance from Site 1','SNAP2_score','Site2_dist','Tempfactor']

features = pd.read_csv('KatG_mutations_df.csv')
```

```python
resid = []
for i in features['KatG mutations']:
    a = i[1:]
    b = a[:-1]
    resid.append(b)
features['resid'] = resid
```

```python
unique_features = features.drop_duplicates(subset='resid',keep = 'first')
unique_features.dropna(inplace = True)
unique_features['Distance from Site 1'] = \
unique_features['Distance from Site 1'].astype(float, errors = 'raise')
unique_features.reset_index(inplace = True)
unique_features.drop(['index','Unnamed: 0'], inplace = True, axis = 1)
#unique_features.drop(unique_features[unique_features['Distance from Site 1'] == 'No centre of mass'], inplace = True)
```

```python
unique_features
```

```python
for a,b in enumerate(titles):
    y = str(column[a])
    print(np.mean(unique_features[y]))
```

```python
for a,b in enumerate(titles):
    y = str(column[a])
    KatG.select_atoms('all').tempfactors = np.mean(unique_features[y])
    with MDAnalysis.Writer(b, multiframe=True, bonds=None, n_atoms=KatG.atoms.n_atoms) as PDB:  
        col_no = unique_features.columns.get_loc(y)

        for i in unique_features['resid']:
            res = 'resid ' + i
            residue = KatG.select_atoms(res)
            row_no = unique_features[unique_features['resid'] == i].index[0]
            feature_value = unique_features.iloc[row_no, col_no]
            residue.tempfactors = feature_value
            x = np.nan_to_num(KatG.select_atoms('all').tempfactors, copy=True, \
                              nan=np.mean(unique_features[y]))
            KatG.select_atoms('all').tempfactors = x
        PDB.write(KatG.atoms)
```

Code to copy and paste into VMD to highlight resistant mutations

```python
ML_data = pd.read_csv('ML_data.csv')
```

```python
ML_data[ML_data['MUTATION']=='S315T']
```

```python
ML_data_unique = ML_data.drop_duplicates(subset='MUTATION', keep = 'first')
ML_data_unique.reset_index(inplace = True)
ML_data_unique.drop(['index','Unnamed: 0'], inplace = True, axis = 1)
```

```python
resid = []
for i in ML_data_unique['MUTATION']:
    a = i[1:]
    b = a[:-1]
    resid.append(b)
ML_data_unique['resid'] = resid
```

```python
resistant = []
susceptible = []

col_no = ML_data_unique.columns.get_loc('MUT_RESISTANCE')

for i in ML_data_unique['resid']:
    row_no = ML_data_unique[ML_data_unique['resid'] == i].index[0]
    RvS = ML_data_unique.iloc[row_no, col_no]
    if RvS == 1:
        susceptible.append(i)
    elif RvS == 0:
        resistant.append(i)
mixed = np.intersect1d(resistant, susceptible)
```

```python
print('Number of residues with resistant mutations: ',(len(resistant)))
print('Number of residues with susceptible mutations: ',(len(susceptible)))
print('Number of residues with mixed mutations: ',(len(mixed)))
```

Printing lists of residues associated with each DSSP secondary structure region

```python
all_features = pd.read_csv('KatG_mutations_df.csv')

resid = []
for i in all_features['KatG mutations']:
    a = i[1:]
    b = a[:-1]
    resid.append(b)
all_features['resid'] = resid

unique_resid_features = all_features.drop_duplicates(subset='resid',keep = 'first')
unique_resid_features.reset_index(inplace = True)
```

```python
#create empty arrays associated with each ss
H = []
O = []
T = []
S = []
G = []
E = []
B = []
I = []
ss_list = [H,O,T,S,G,E,B,I]
ss_string_list = ['H','O','T','S','G','E','B','I']

#sets column no for resid number
col_no = unique_resid_features.columns.get_loc('resid')

#cycles through each row for each ss, appending the respective resid with the correct ss feature
for number,item in enumerate(ss_string_list):
    row_no = -1
    for i in unique_resid_features[item]:
        row_no += 1
        if i == 1:
            (ss_list[number]).append(unique_resid_features.iloc[row_no,col_no])
            
            
```
